# Data Import

```python
import pandas as pd
import numpy as np
import os
from glob import glob #returns a list of files matching a given pattern
import sqlite3
conn = sqlite3.connect('movies_db.sqlite')
cur = conn.cursor()
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import a list of all files from the zipped folder
all_files = glob('./zippedData/*.csv.gz') 
all_files
```

```python
#create an empty dictionary to store the data
dict_files = {}

#for loop to populate the dictionary
for file in all_files:
    dict_files[file] = pd.read_csv(file)
```

```python
dict_files.keys() 
```

```python
#clean the file names to a common format
cleaned_files = {}
for filename in all_files:
    csv_cleaning = os.path.basename(filename).replace('.csv', '').replace('.', '_')
    filename_df = pd.read_csv(filename, index_col=0)
    cleaned_files[csv_cleaning] = filename_df
```

```python
cleaned_files.keys()
```

```python
def create_sql_table_from_df(df, name, conn):
    try:
        df.to_sql(name, conn)
        print(f"Created table {name}")
    
    except Exception as e:
        print(f"could not make table {name}")
        print(e)
```

```python
for name, table in cleaned_files.items():
    create_sql_table_from_df(table, name, conn)
```

```python
for item in cleaned_files.items():
    print('------------------')
    print(item[0])
    print(item[1].info())
    print('------------------')
```

# QUESTION 1 - How much do the top 25 profitable movies compare pre and post inflation adjusted?

## Clean Budgets Dataframe

```python
#Convert sequel table to pandas df
cur.execute("""SELECT * FROM tn_movie_budgets_gz""")

budgets_df = pd.DataFrame(cur.fetchall())
budgets_df.columns = [x[0] for x in cur.description]
budgets_df.head()
```

```python
budgets_df.info()
```

```python
# remove symbols and convert datatype to int for multiple columns
def int_conversion(df, col):
    df[col] = df[col].str.replace('$', '').str.replace(',', '').astype('float')
    return df
```

```python
convert_cols = ['production_budget', 'domestic_gross', 'worldwide_gross']

for col in convert_cols:
    budgets_df = int_conversion(budgets_df, col)
```

```python
budgets_df.info()
```

```python
#view top ten highest grossing films pre cpi adjustment
budgets_df.sort_values(by='worldwide_gross', ascending=False, inplace=True)
```

```python
#convert release date from object to datetime datatype
budgets_df['release_date'] = pd.to_datetime(budgets_df['release_date'])
budgets_df.head()
```

```python
budgets_df.info()
```

```python
#Calculate net profit
budgets_df['gross_profit_loss'] = budgets_df['worldwide_gross'] - budgets_df['production_budget']
budgets_df.head()
```

### Top 25 Profitable Films pre Inflation Adjusted

```python
#sort movie net profit by most profitable
budgets_df.sort_values(by='gross_profit_loss', ascending=False, inplace=True)
```

```python
top_25_movies_pre_df = budgets_df[:25]
```

```python
top_25_movies_pre_df
```

#### Graph Results

```python

sns.set_style('darkgrid')
barplot = sns.barplot(x ='gross_profit_loss', y ='movie', 
                      data= top_25_movies_pre_df, color='green')
barplot.set(xlabel ='Profit (Billions)', ylabel ='Movie')
sns.set(rc={'figure.figsize':(20,15)})
plt.title('Pre-Inflation Adjusted Top 25 Most Profitable Films', fontsize=20)
plt.show()
```

### Top 25 Profitable Films Adjusted

```python
#break out release date by year
budgets_df['year'] = budgets_df['release_date'].dt.year
```

```python
budgets_df.head()
```

```python
#create new df with only 2019 released movies
budget_2019 = budgets_df[budgets_df.year == 2019]
```

```python
budget_2019.info()
```

```python
#remove 2019 and 2020, because CPI module doesn't support those years
budget_df_2018 = budgets_df.drop(budgets_df[(budgets_df.year == 2019) & 
                                                      (budgets_df.year == 2020)].index) #what is .index
```

```python
budget_df_2018 = budget_df_2018[budget_df_2018.year != 2020]
```

```python
budget_df_2018.year.unique()
```

#### Import CPI module

```python
!pip install cpi
import cpi
cpi.update()
```

```python
#run CPI module to adjust net profit for inflation
budget_df_2018['worldwide_adjusted'] = budget_df_2018.apply(lambda x: 
                                        cpi.inflate(x.gross_profit_loss, x.year), axis=1)
```

```python
budget_df_2018.info()
```

```python
#Add 2019 values back into df
budgets_adjusted_df = budget_df_2018.append(budget_2019)
```

```python
budgets_adjusted_df.info()
```

```python
#fill inflation adjusted 2019 null values with pre-adjusted numbers
budgets_adjusted_df['worldwide_adjusted'].fillna(budgets_adjusted_df['gross_profit_loss'], inplace=True)
```

```python
#check work
budgets_adjusted_df.loc[budgets_adjusted_df['year'] == 2019]
```

```python
budgets_adjusted_df.info()
```

```python
budgets_adjusted_df.head()
```

```python
budgets_adjusted_df.sort_values(by='worldwide_adjusted', ascending=False,
                               inplace=True)
budgets_adjusted_df.head()
```

#### Graph Results

```python
top_25_adjusted = budgets_adjusted_df[:25]
top_25_adjusted
```

```python
sns.set_style('darkgrid')
barplot = sns.barplot(x ='worldwide_adjusted', y ='movie', data= top_25_adjusted, color='green')
barplot.set(xlabel ='Profit (Billions)', ylabel ='Movie')
sns.set(rc={'figure.figsize':(20,15)})
plt.title('Inflation Adjusted Top 25 Most Profitable Films', fontsize=20)
plt.show()
```

```python
sns.set_style('darkgrid')
barplot = sns.barplot(x ='worldwide_adjusted', y ='movie', data= top_25_adjusted, color='green')
barplot.set(xlabel ='Profit (Billions)', ylabel ='Movie')
plt.show()
```

## QUESTION 1.1 - Is there a correlation between ROI and Overall Profit for the top 25 ROI movies?

```python
#Create ROI column. Dealing with a %, so inflation adjusted shouldn't matter
budgets_adjusted_df['ROI'] = (budgets_adjusted_df['worldwide_gross'] - 
                                budgets_adjusted_df['production_budget']) / budgets_adjusted_df['production_budget']
```

```python
budgets_adjusted_df.head()
```

```python
top_25_movies_roi_df = budgets_adjusted_df[:25]
```

```python
top_25_movies_roi_df.head(25)
```

### Graph Results

```python
sns.set_style('darkgrid')
barplot = sns.barplot(x ='ROI', y ='movie', data= top_25_movies_roi_df, color='green')
barplot.set(xlabel ='ROI', ylabel ='Movie')
sns.set(rc={'figure.figsize':(20,15)})
plt.title('ROI(%) vs Top 25 Most Profitable Films', fontsize=20)
plt.show()
```

## Conclusions and Recommendations

There's a clear difference between the top 25 films when accounting for inflation. The number one most profitable film of all time is "Gone with the Wind" at over seven billion dollars. "Avatar", the number one film without adjusting profits for inflation drops down to sixth overall. 

Visualising the ROI of the top 25 most profitable films provides a clear view of how production budgets can impact investment returns. For example, Avatar's net profit of $2.8 billion only netted investors a 5% ROI. Where as the movie "Bambi", the second most profitable film, had a much smaller production budget, but returned 311% to investors and the company.

When exploring movie data for profitability in any capacity, it's highly recommended to adjust the financial information for inflation, which will provide much more accurate conclusions for future investment considerations.


# QUESTION 2 - Most Profitable Films by Release Month

```python
#Create column for release month
budgets_adjusted_df['month'] = budgets_df['release_date'].dt.month
```

```python
budgets_adjusted_df.head()
```

```python
top_200_films = budgets_adjusted_df
```

```python
#Sort the top 200 films by net profit
top_200_films.sort_values(by='worldwide_adjusted', ascending = False, inplace = True)
top_200_films.head()
```

```python
top_200_films_final_df = top_200_films[:200]
```

```python
#Group the sum of net profit by release month
top_200_profit_month = top_200_films_final_df.groupby(['month'])['worldwide_adjusted'].sum()
```

```python
type(top_200_profit_month)
```

```python
top_200_profit_month
```

## Graph Results of top 200 Movies by Release Month

```python
top_200_profit_month.plot(kind='line', use_index=True)

plt.title('Top 200 Movies Profit by Release Month', fontsize=20)
plt.xlabel('Month')
plt.ylabel('Profit (Billions*10)')
```

```python
sns.set(style='ticks')

f, ax = plt.subplots(figsize = (10,10))


sns.boxplot(x = 'month', y = 'worldwide_adjusted', data = top_200_films_final_df,
           color='green')
plt.title('Top 200 Movie Profits by Release Month', fontsize=20)
plt.xlabel('Month')
plt.ylabel('Profit (Billions)')
plt.show()
```

## Graph Top 200 Films by Release Month Average

```python
#Average the profit by month
top_200_profit_month_avg = top_200_films_final_df.groupby(['month'])['worldwide_adjusted'].mean()
```

```python
top_200_profit_month_avg.plot(kind='line', use_index=True)

plt.title('Top 200 Movies Average Profit by Release Month', fontsize=20)
plt.xlabel('Month')
plt.ylabel('Averaged Profit (Billions)')
```

## Conclusions and Recommendations

The most profitable times to release a movie, based on the top 200 films, are during late spring/early summer, and around the holday months. December is the most profitable by half a billion dollars over the second most profitable month, May. The least profitable time to release a new film is after the holidays or during late summer(September). This could be due to the fact that moviegoers have more time to see movies during the profitable months, and less time during the least profitable because of school and/or work.

When looking at the average profit by month, the most profitable release time is around the holidays (December), and the least profitable months stay the same. However, a few things change when looking at average profit in this way. First, the  beginning summer months don't look as enticing. The average profit by film is most likely pulled down by the sheer number of movies released during these months with lower overall profits. second, there is a huge spike in August due to a single outlier pulling up the average profit for this month.

The data shows that the best time to release a film for the highest potential realized profit is during December, and the worst time to release a film is in January. 

# QUESTION 3 - Does Rating Correlate with Profitability?

### Data Import and Clean for Question 3

#### Genres and Runtime df

```python
cur.execute("""SELECT * FROM imdb_title_basics_gz""")

genres_df = pd.DataFrame(cur.fetchall())
genres_df.columns = [x[0] for x in cur.description]
genres_df.head()
```

```python
genres_df.info()
```

```python
genres_df = genres_df.drop('start_year',  axis=1)
```

```python
genres_df = genres_df.drop('original_title',  axis=1)
```

```python
genres_df.info()
```

#### Ratings df

```python
cur.execute("""SELECT * FROM imdb_title_ratings_gz""")

ratings_df = pd.DataFrame(cur.fetchall())
ratings_df.columns = [x[0] for x in cur.description]
ratings_df.head()
```

```python
ratings_df.info()
```

##### Merge Ratings and Genres df

```python
movie_ratings_df = pd.merge(genres_df, ratings_df, 
                                 left_on = ['tconst'], right_on = ['tconst'],
                                 how = 'right')
movie_ratings_df.head()
```

```python
movie_ratings_df.info()
```

```python
budgets_adjusted_df.info()
```

```python
#Merge dataframes on the common movies from each.
profit_ratings_df = pd.merge(budgets_adjusted_df, movie_ratings_df, 
                                 left_on = ['movie'], right_on = ['primary_title'],
                                 how = 'inner')
profit_ratings_df.head()
```

```python
profit_ratings_df.info()
```

```python
profit_ratings_df.drop_duplicates(inplace=True)
```

```python
profit_ratings_df.info()
```

## Graph Results

```python
sns.jointplot(x='averagerating', y='worldwide_adjusted', height=15, 
              data=profit_ratings_df, kind='reg', truncate=True,
             xlim=(1.5,9.5))
plt.title('Rating vs Profitability', fontsize=20)
plt.xlabel('Rating')
plt.ylabel('Profit (Billions)')
```

```python
profit_vs_rating = profit_ratings_df.groupby(['averagerating'])['worldwide_adjusted'].mean()
```

```python
profit_vs_rating
```

```python
profit_vs_rating.plot(kind='line', use_index=True)

plt.title('Rating vs Profit Averaged', fontsize=20)
plt.xlabel('Profit')
plt.ylabel('Averaged Profit (Billions)')
```

## Conclusions and Recommendations

There is a wide range of profitability based on movie rating. Many highly rated movies have generated negative income. Conversely, there are profitable films that have been rated pooly as well. However, there seems to be a slight positive correlation between rating and profit. This becomes particularly clear when averaging the profit by rating. The most profitable movies by rating lie within the 7-8 range. Lack of data point desnity, along with a few outliers creat false positive results between the rating of 2 and 3.

The most profitable movies are the most highly rated. In order to generate the most positive cash flow on a film, it recommended to fall between a 7-8 rating. This can be achieved through market research and testing the movie concept before production and pre-release.
